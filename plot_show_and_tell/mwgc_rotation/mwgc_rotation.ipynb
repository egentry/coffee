```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
mpl.rc('figure', figsize=(12, 9))
mpl.rc('font', size=20)
```

```python
pos_df = pd.read_fwf('harris2010/mwgc_pos.dat')
vel_df = pd.read_fwf('harris2010/mwgc_vel.dat')
feh_df = pd.read_fwf('harris2010/mwgc_feh.dat')

L, B, R = pos_df.L.values, pos_df.B.values, pos_df.R_gc.values
X, Y, Z = pos_df.X.values, pos_df.Y.values, pos_df.Z.values

v_r, v_LSR = vel_df.v_r.values, vel_df.v_LSR.values

feh = feh_df['[Fe/H]'].values
ub = feh_df['U-B'].values
bv = feh_df['B-V'].values
vr = feh_df['V-R'].values
vi = feh_df['V-I'].values
```

```python
plt.scatter((L + 180) % 360 - 180, B, c=v_LSR, cmap='coolwarm', s=100, alpha=0.8, edgecolor='k')
plt.xlabel("$\Delta \ell$  [deg]")
plt.ylabel("$b$  [deg]")
cb = plt.colorbar(extend='both')
cb.set_label("$v_\mathrm{LSR}$  [km s$^{-1}$]")

# for spine removal (less gruesome than it sounds) 
# http://matplotlib.org/examples/ticks_and_spines/spines_demo.html
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
```
